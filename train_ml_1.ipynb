```python
x_tr, x_te, y_tr, y_te= train_test_split(x_diabetes, y_diabetes, train_size=0.7, random_state=42)
```

```python
modeld=linear_model.LinearRegression().fit(x_tr, y_tr)
print(modeld.coef_)
print(modeld.intercept_)
r2=modeld.score(x_tr,y_tr)

y_pr=modeld.predict(x_tr)
sse_y_pr= np.sum((y_tr - y_pr)**2 ) / y_pr.size

y_pr_te= modeld.predict(x_te)
sse_y_te= np.sum(( y_te - y_pr_te)**2 ) / y_pr_te.size

print(sse_y_pr , sse_y_te)

```

```python
modelr= linear_model.Ridge(alpha=0.1).fit(x_tr, y_tr)
y_tr_pr_r=modelr.predict(x_tr)
sse_y_prr= np.sum((y_tr - y_tr_pr_r)**2 )/(y_tr_pr_r.size)

y_te_r= modelr.predict(x_te)
sse_y_ter= np.sum(( y_te - y_te_r)**2 ) / (y_te_r.size)
print(sse_y_prr, sse_y_ter)
```
